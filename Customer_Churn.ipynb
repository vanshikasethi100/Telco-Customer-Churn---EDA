```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customer Churn analysis.csv')
df.head()
```

```python
df.info()
```

```python
#replacing blanks with as tenure is 0 and no total charges are recorded.
#also changed the datatype to float
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df["customerID"].duplicated().sum()
```

```python
#converted 0 oand 1 value of seniorcitizen to yes and no
def conv(value):
    if value==1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.head()
```

```python
#Churn count
ax = sns.countplot(x = df['Churn'])
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")

plt.title("Percentage of Churned Customers", fontsize=9)
plt.show()
```

#from the given pie chart we can conclude that 26.54% of customers have churned out. lets explore the reason behind it.

```python
plt.figure(figsize=(3,4))
ax = sns.countplot(x = 'gender', hue = "Churn", data=df)
ax.bar_label(ax.containers[0])
plt.title("Churn by Gender")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---- groupby counts ----
g = df.groupby(['SeniorCitizen','Churn']).size().reset_index(name='count')

# pivot
pivot = g.pivot(index='SeniorCitizen', columns='Churn', values='count').fillna(0)

# ---- row percentage (REAL churn rate) ----
percent = pivot.div(pivot.sum(axis=1), axis=0) * 100

# ---- plot ----
ax = percent.plot(kind='bar',
                  stacked=True,
                  figsize=(5,4),
                  color=['#1f77b4', '#ff7f0e'])   # blue & orange

plt.title("Churn Rate by SeniorCitizen")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage")

# ---- add labels ----
for container in ax.containers:
    labels = [f"{bar.get_height():.1f}%" if bar.get_height() > 0 else "" for bar in container]
    ax.bar_label(container, labels=labels, label_type='center', fontsize=9, color='black')

plt.legend(title='Churn')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

#comparatively greater % of people in senior citizens have churned out

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(x = 'Contract', data=df, hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Contract")
plt.show()
```

#people who have month to month contract are more likely to churn and people who contract upto 1 or 2 years are less likely to churn

```python
cat_cols = [
'PhoneService','MultipleLines',
'InternetService','OnlineSecurity','OnlineBackup','DeviceProtection',
'TechSupport','StreamingTV','StreamingMovies','PaperlessBilling',
'PaymentMethod'
]

sns.set_style("whitegrid")

n_cols = 3
n_rows = (len(cat_cols) + n_cols - 1) // n_cols

plt.figure(figsize=(18, 5*n_rows))

for i, col in enumerate(cat_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.countplot(data=df, x=col, color='#1f77b4')

    plt.title(col)
    plt.xticks(rotation=5, ha='center')   # ← horizontal labels
    plt.xlabel('')
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
